### Dealing with Null Values

```python
def fillNaMean(df, numeric_cols, categorical_cols):
    for col in numeric_cols:
        df[col] = df[col].fillna(df[col].mean())
    for col in categorical_cols:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
numeric_columns = ['Final_Score', 'Assignments_Avg', 'Quizzes_Avg', 'Participation_Score', 'Projects_Score', 'Total_Score']
categorical_columns = ['Grade', 'Parent_Education_Level']
fillNaMean(df, numeric_columns, categorical_columns)
```

### Dropping Irrelevant Columns

```python
df.drop('Student_ID', axis=1, inplace=True)
df.drop('First_Name', axis=1, inplace=True)
df.drop('Last_Name', axis=1, inplace=True)
df.drop('Email', axis=1, inplace=True)
df.drop('Gender', axis=1, inplace=True)
df.drop('Department', axis=1, inplace=True)
```

```python
df.info()
```

```python
def changeFloattoInt64(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Attendance (%)', 'Total_Score', 'Study_Hours_per_Week', 'Sleep_Hours_per_Night', 'Projects_Score', 'Participation_Score', 'Quizzes_Avg', 'Assignments_Avg', 'Final_Score', 'Midterm_Score']
changeFloattoInt64(columns)
```

```python
df.info()
```

```python
df_clean = df.copy()
```

```python
def encodeCols(cols):
    for i in cols:
        temp = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(temp))
        df_clean[i] = data_LE.transform(df[i])

columns = ['Family_Income_Level', 'Parent_Education_Level', 'Extracurricular_Activities','Internet_Access_at_Home','Grade']
encodeCols(columns)
```

### Data Splitting / Train-Test Split

```python
X = df_clean.drop(columns=['Grade'])
y = df_clean['Grade']

trainX, testX, trainY, testY = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
trainX.head(2)
```

```python
trainY.head(2)
```

```python
print("TrainX:",trainX.shape)
print("TrainY:",trainY.shape)
print("TestX:",testX.shape)
print("TestY:",testY.shape)
```

### Training / Applying Classifier

```python
import pickle
from sklearn.svm import SVC
```

```python
model_SVC = SVC()
model_SVC.fit(trainX, trainY)

print(model_SVC)
```

### Testing & Processing Results

```python
svc_predictions = model_SVC.predict(testX)
```

```python
print(svc_predictions)
```

### Accuracy Score

```python
from sklearn.metrics import accuracy_score
svc_accuracy = accuracy_score(testY, svc_predictions)

print("-- Model Accuracy: ", end="")
print(round(svc_accuracy,3))
```
